### SCATTER PLOTS:

```python
# Scatter plot of sepal length v/s sepal width
sns.scatterplot(x='sepal length', y=' sepal width', data=iris, hue=' class')
plt.title('Scatter Plot: Sepal Length vs Sepal Width')
plt.show()
```

```python
# Scatter plot of petal length v/s petal width
sns.scatterplot(x=' petal length', y=' petal width', data=iris, hue=' class')
plt.title('Scatter Plot: Petal Length vs Petal Width')
plt.show()
```

### BOX PLOTS:

```python
# Box plot of sepal length for each class
sns.boxplot(x=' class', y='sepal length', data=iris)
plt.title('Box Plot: Sepal Length by Class')
plt.show()
```

```python
# Box plot of sepal width for each class
sns.boxplot(x=' class', y=' sepal width', data=iris)
plt.title('Box Plot: Sepal Width by Class')
plt.show()
```

```python
# Box plot of Petal length for each class
sns.boxplot(x=' class', y=' petal length', data=iris)
plt.title('Box Plot: Petal Length by Class')
plt.show()
```

```python
# Box plot of petal width for each class
sns.boxplot(x=' class', y=' petal width', data=iris)
plt.title('Box Plot: Petal Width by Class')
plt.show()
```

### DENSITY PLOTS:

```python
# Density plot  of Petal width
sns.kdeplot(data=iris, x=' petal width', hue=' class', fill=True)
plt.title('Density Plot - Petal Width')
plt.xlabel('Petal Width')
plt.ylabel('Density')
plt.show()
```

```python
# Density plot  of Petal Length
sns.kdeplot(data=iris, x=' petal length', hue=' class', fill=True)
plt.title('Density Plot - Petal Length')
plt.xlabel('Petal Length')
plt.ylabel('Density')
plt.show()
```

```python
# Density plot  of Sepal width
sns.kdeplot(data=iris, x=' sepal width', hue=' class', fill=True)
plt.title('Density Plot - Sepal Width')
plt.xlabel('Sepal Width')
plt.ylabel('Density')
plt.show()
```

```python
# Density plot  of Sepal Length
sns.kdeplot(data=iris, x='sepal length', hue=' class', fill=True)
plt.title('Density Plot - Sepal Length')
plt.xlabel('Sepal Length')
plt.ylabel('Density')
plt.show()
```

### PAIRPLOTS:

```python
# Pairplot to visualize relationships between different features
sns.pairplot(data=iris, hue=' class')
plt.title('Pairplot: Iris Dataset')
plt.show()
```

### VIOLIN PLOTS:

```python
# Violin plot of sepal lenght for each class
sns.violinplot(x=' class', y='sepal length', data=iris)
plt.title('Violin Plot: Sepal Length by Class')
plt.show()
```

```python
# Violin plot of sepal width for each class
sns.violinplot(x=' class', y=' sepal width', data=iris)
plt.title('Violin Plot: Sepal Width by Class')
plt.show()
```

```python
# Violin plot of petal length for each class
sns.violinplot(x=' class', y=' petal length', data=iris)
plt.title('Violin Plot: Petal Length by Class')
plt.show()
```

```python
# Violin plot of petal width for each class
sns.violinplot(x=' class', y=' petal width', data=iris)
plt.title('Violin Plot: Petal Width by Class')
plt.show()
```
